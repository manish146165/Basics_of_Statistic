# Mean, Median, Mode & Standard Deviation

# Mean of the scores.

```python
import statistics
import matplotlib.pyplot as plt

name=['Sonu','Himanshu','Samarth','Dhiraj','Manish','Rahul','Deepak','Praveen','Ankur','Amit','Kavita','Arjun']

Task_Scores=[3.9,3.5,3.5,3.4,3.4,3.4,3.3,3.3,3.21,3.158,3,1.72]

# calculate the mean
mean=statistics.mean(Task_Scores)
print('Mean of the Task Scores=',mean)

# plot bar graph 
plt.bar(name,Task_Scores)
plt.xlabel("Name of the interns")
plt.ylabel('Task Scores')
plt.title('Marks of the student')
plt.xticks(rotation=65)
plt.ylim(0,5)
plt.axhline(y=mean, color='r', linestyle='--', label='Mean =3.23')
plt.legend()
plt.show() 
print('Mean of the Task Scores=',mean)
```

# Median of the scores.

```python
import statistics
import matplotlib.pyplot as plt

name=['Sonu','Himanshu','Samarth','Dhiraj','Manish','Rahul','Deepak','Praveen','Ankur','Amit','Kavita','Arjun']

Task_scores=[3.9,3.5,3.5,3.4,3.4,3.4,3.3,3.3,3.21,3.158,3,1.72]

# calculate the mean
median=statistics.median(Task_scores)
print('Median of the Task Scores=',median)

# plot bar graph 
plt.bar(name,Task_scores)
plt.xlabel("Name of the interns")
plt.ylabel('Task scores')
plt.title('Marks of the interns')
plt.xticks(rotation=65)
plt.ylim(0,5)
plt.axhline(y=median, color='r', linestyle='--', label='Median =3.34')
plt.legend()
plt.show()
print('Median of the Task Scores=',median)
```

# Mode of the scores.

```python
import statistics
import matplotlib.pyplot as plt

name=['Sonu','Himanshu','Samarth','Dhiraj','Manish','Rahul','Deepak','Praveen','Ankur','Amit','Kavita','Arjun']

Task_scores=[3.9,3.5,3.5,3.4,3.4,3.4,3.3,3.3,3.21,3.158,3,1.72]

# calculate the mean
mode=statistics.mode(Task_scores)
print('Mode of the Task Scores=',mode)

# plot bar graph 
plt.bar(name,Task_scores)
plt.xlabel("Name of the interns")
plt.ylabel('Task_scores')
plt.title('Marks of the interns')
plt.xticks(rotation=65)
plt.ylim(0,5)
plt.axhline(y=mode, color='r', linestyle='--', label='Mode =3.4')
plt.legend()
plt.show() 
print('Mode of the Task Scores',mode)
```

# Standard Deviation of the scores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data and names
names = ['Sonu','Himanshu','Samarth','Dhiraj','Manish','Rahul','Deepak','Praveen','Ankur','Amit','Kavita','Arjun']
Task_Scores = [3.9, 3.5, 3.5, 3.4, 3.4, 3.4, 3.3, 3.3, 3.21, 3.158, 3, 1.72]

# Calculate the standard deviation and mean of the data
std_dev = np.std(Task_Scores)
print('Standard deviation of the Task Scores=',std_dev)
mean = np.mean(Task_Scores)
print('Mean of the Task Scores=',mean)

print('1.Eight Interns (Manish,Sonu,Himanshu,Deepak,Rahul,Dhiraj,Samarth,Preveen),has scores larger than the average      scores.')
print('2.Four Interns (Arjun,Amit,Kavita,Ankur) have scores less than the average scores.')
print('3.Arjun and Sonu both fall within a high range of Standard Deviation.')
print('4.Manish, Dhiraj, Himanshu, Kavita, Samarth, Amit, Ankur, Deepak, Praveen, and Rahul all fall within the low             standard deviation range.')


# Create a scatter plot
plt.axvline(mean, color='r', linestyle='--', linewidth=1, label='Mean')
plt.axvline(x=mean - std_dev, color='g', linestyle='--', label='Mean - Std Dev')
plt.axvline(x=mean + std_dev, color='black', linestyle='--', label='Mean + Std Dev')
plt.scatter(Task_Scores,names, color='b', label='Task Scores')


# Add labels and title
plt.xlabel('Task Scores')
plt.ylabel('Names of the interns')

plt.xticks(rotation=65)
plt.title('Standard Deviation with Mean')

# Add a legend
plt.legend()

# Show the plot
plt.show()
print('Standard deviation of the Task Scores=',std_dev)
print('Mean of the Task Scores=',mean)
```
